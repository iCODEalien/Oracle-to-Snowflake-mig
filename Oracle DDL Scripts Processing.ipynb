```python
# creating a function to pass prompt and get responses from LLM

def GetResponse(prompt, model = 'gpt-3.5-turbo'):
    # message to the model
    messages = [{'role': 'user', 'content':prompt}]
    
    # response from the model
    response = openai.ChatCompletion.create(messages = messages,
                                             model = model,
                                             temperature = 0)
    return response.choices[0].message['content']
```

```python
# Path where LLM translated scripts are available.
AllDDL_path = r'C:\Users\Chaitra.b.c\Documents\Oracle to Snowflake Migration'
SnowDDL_path = '\\Snowflake_DDL_scripts'

# Now we need to create a folder if not there, create
def createSnowDDL_dir(folder_name):
    try:
        # if the folder exists...
        os.chdir(AllDDL_path + SnowDDL_path + '\\' + folder_name)
        folder_path = AllDDL_path + SnowDDL_path + '\\' + folder_name
        print('Directory changed to ',folder_name)
        return folder_path
    # else create new one and go to that directory
    except:
        if os.path.exists(AllDDL_path + SnowDDL_path):
            os.mkdir(AllDDL_path + SnowDDL_path + '\\' + folder_name)
            os.chdir(AllDDL_path + SnowDDL_path + '\\' + folder_name)
        else:
            os.mkdir(AllDDL_path + SnowDDL_path)
            os.chdir(AllDDL_path + SnowDDL_path)
            os.mkdir(AllDDL_path + SnowDDL_path + '\\' + folder_name)
            os.chdir(AllDDL_path + SnowDDL_path + '\\' + folder_name)
        folder_path = AllDDL_path + SnowDDL_path + '\\' + folder_name
        print('New Directory Created...',folder_name,' and navigated to')
        return folder_path
```

```python
# Oracle to Snowflake Translations.

for folder in list_of_folders:
    # Considering only Folders
    if '.sql' not in folder:
        folder_path = path + '\\' + folder
        print('Folder: ',folder_path)
        
        # Changing path to that folder
        os.chdir(folder_path)
        
        # Storing the list of files in that path
        list_of_files = os.listdir()
        
        # Iterating through each file
        for file in list_of_files:
            file_path = folder_path + '\\' + file
            print('File: ',file_path)
            print('*'*50)
            
            # opening the current file for read
            with open(file_path,'r') as f:
                #print(f.readlines())
                #print(f.read())
                DDL_statement = f.read()
                print('Oracle Syntax: \n', DDL_statement)
                
                # creating prompt for LLM.
                prompt = "'" + DDL_statement + "'"+ '''
                Convert the Data Types of this Oracle DDL statement to Snowflake Data Types. 
                If Oracle's data types are compatible with Snowflake's Data types then 
                just return the original string 
                Else, just give me the Snowflake DDL statement and don't include Notes, considerations, 'No conversion needed..' stuff or special mention in the output.
                '''
                print('Prompt : \n',prompt,'\n')
                
                # creating a response string and passing the prompt
                response = ''
                response = GetResponse(prompt)
                
                # Storing Response
                if response:
                    #print('Snowflake Syntax : \n',response)
                    print('*'*100)
                    # creating a folder name and navigate to it ,if not exists,
                    # navigate to it and write into it.
                    Snowfolder_path = createSnowDDL_dir(folder)
                    print(Snowfolder_path)
                    
                    # Writing into the file
                    with open(Snowfolder_path + '\\' + file , 'w+') as sqlfile:
                        sqlfile.write(response)
                        
                    # sleep 30s to avoud Rate Limit Error.
                    time.sleep(30)
                    
                else:
                    # sleep 20s
                    time.sleep(20)
                    print('*'*100)
            print('*'*100)
    
```
